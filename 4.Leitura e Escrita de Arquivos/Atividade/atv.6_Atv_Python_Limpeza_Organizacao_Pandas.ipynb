## 游닇 Exerc칤cio 3
Com o arquivo `financeiro.csv`:

a) Renomeie `m칡s` para `mes`

b) Crie coluna `resultado` (receita - despesa)

c) Calcule o saldo acumulado

```python
#carregando a planillha
import pandas as pd
dados = pd.read_csv("/content/financeiro.csv", sep=";", encoding="latin")
dados.head()

#a) renomeando "m칡s" para "mes"
dados = dados.rename(columns={"m칡s":"mes"})
dados.head()
```

```python
#b) Crie coluna `resultado` (receita - despesa)
dados['resultado'] = dados['receita'] - dados['despesa']
dados.head()
```

```python
#c) Calcule o saldo acumulado
dados['saldo_acumulado'] = dados['resultado'].cumsum()
dados.head()
```

## 游빌 Tarefa para Casa
Utilize o arquivo `financeiro2.csv`:
1. Corrija o nome da coluna `m칡s` para `mes`.
2. Preencha valores ausentes com 0.
3. Crie a coluna `resultado` = receita - despesa.
4. Crie a coluna `situacao`, onde:
   - 랇ucro se `resultado > 0`
   - 랋reju칤zo se `resultado < 0`
   - 란quil칤brio se `resultado == 0`
5. Exiba o total de meses com lucro, preju칤zo e equil칤brio.

```python
# 1) corrigindo o nome da coluna "m칡s" para "mes"
import pandas as pd
dados = pd.read_csv("/content/financeiro2.csv", sep=";")


dados = dados.rename(columns={"m칡s":"mes"})
dados.head()
```

```python
#2) Preencha valores ausentes com 0.
dados = dados.fillna(0)
dados
```

```python
#3)Crie a coluna `resultado` = receita - despesa.
dados['receita'] = dados['receita'].astype(str).str.replace(",", ".").astype(float)
dados['resultado'] = dados['receita'] - dados['despesa']
display(dados.head())
```

```python
# 4) Crie a coluna 'situa칞칚o'
def get_situacao(resultado):
  if resultado > 0:
    return 'lucro'
  elif resultado < 0:
    return 'preju칤zo'
  else:
    return 'equilibrio'

dados['situacao'] = dados['resultado'].apply(get_situacao)
display(dados.head())
```

```python
#5)Exiba o total de meses com lucro, preju칤zo e equil칤brio.
situacao_counts = dados['situacao'].value_counts()
print(situacao_counts)
```
