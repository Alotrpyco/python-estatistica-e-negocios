## üìù Exerc√≠cio 3
Com o arquivo `financeiro.csv`:

a) Renomeie `m√™s` para `mes`

b) Crie coluna `resultado` (receita - despesa)

c) Calcule o saldo acumulado

```python
#carregando a planillha
import pandas as pd
dados = pd.read_csv("/content/financeiro.csv", sep=";", encoding="latin")
dados.head()

#a) renomeando "m√™s" para "mes"
dados = dados.rename(columns={"m√™s":"mes"})
dados.head()
```

```python
#b) Crie coluna `resultado` (receita - despesa)
dados['resultado'] = dados['receita'] - dados['despesa']
dados.head()
```

```python
#c) Calcule o saldo acumulado
dados['saldo_acumulado'] = dados['resultado'].cumsum()
dados.head()
```

## üß© Tarefa para Casa
Utilize o arquivo `financeiro2.csv`:
1. Corrija o nome da coluna `m√™s` para `mes`.
2. Preencha valores ausentes com 0.
3. Crie a coluna `resultado` = receita - despesa.
4. Crie a coluna `situacao`, onde:
   - ‚Äúlucro‚Äù se `resultado > 0`
   - ‚Äúpreju√≠zo‚Äù se `resultado < 0`
   - ‚Äúequil√≠brio‚Äù se `resultado == 0`
5. Exiba o total de meses com lucro, preju√≠zo e equil√≠brio.

```python
# 1) corrigindo o nome da coluna "m√™s" para "mes"
import pandas as pd
dados = pd.read_csv("/content/financeiro2.csv", sep=",")


dados = dados.rename(columns={"m√™s":"mes"})
dados.head()
```

```python
#2) Preencha valores ausentes com 0.
import pandas as pd
dados = pd.read_csv("/content/financeiro2.csv", sep=",")
dados = dados.fillna(0)
dados
```

```python
#3)Crie a coluna `resultado` = receita - despesa.
dados['resultado'] = dados['Receita'] - dados['Despesa']
display(dados.head())
```

```python
# 4) Crie a coluna 'situa√ß√£o'
def get_situacao(resultado):
  if resultado > 0:
    return 'lucro'
  elif resultado < 0:
    return 'preju√≠zo'
  else:
    return 'equilibrio'

dados['situacao'] = dados['resultado'].apply(get_situacao)
display(dados.head())
```

```python
#5)Exiba o total de meses com lucro, preju√≠zo e equil√≠brio.
situacao_counts = dados['situacao'].value_counts()
print(situacao_counts)
```
