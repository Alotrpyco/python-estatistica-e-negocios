````python
import pandas as pd
df = pd.read_csv("/content/vendas_ausente.csv", sep = ";", decimal = ",")
df.head()
````

````python
#1) calcular a m√©dia e o desvio padr√£o do pre√ßo unit√°rio
mean = df["preco_unitario"].mean()
print(mean)

std = df["preco_unitario"].std()
print(std)
````

````python
#2) mostrando a mediana de valor_total para cada produto usando o "groupby"
agrupar = df.groupby("produto")["valor_total"].median()
print(agrupar)
````

````python
#3) Exiba qual produto teve maior n√∫mero de vendas (com `value_counts`).
agrupamento = df.groupby("produto")["valor_total"].sum()
decrescente = agrupamento.sort_values(ascending = False)

decrescente.head()
````

````python
#4) Calcule a m√©dia de `quantidade` por m√™s (`data_venda` convertida para m√™s).
df['data_venda'] = pd.to_datetime(df['data_venda'], errors='coerce')
df['mes'] = df['data_venda'].dt.to_period('M')
df.groupby('mes')['quantidade'].mean()
````
