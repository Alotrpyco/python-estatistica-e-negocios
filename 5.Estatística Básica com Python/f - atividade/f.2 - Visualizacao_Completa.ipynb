**Exerc√≠cio 1**  
Crie um gr√°fico de barras com a m√©dia de `preco_unitario` por `produto`.

````python
import pandas as pd
df = pd.read_csv("/content/vendas_ausente.csv", sep=";", decimal = ",")
df.head()
````

````python
#criando um gr√°fico
import matplotlib.pyplot as plt
grupo = df.groupby("produto")["preco_unitario"].mean()
grupo.plot(kind="bar", title = "M√©dia de pre√ßo unit√°rio por produto", color = "black")
plt.xlabel("Produto")
plt.ylabel("Pre√ßo Unit√°rio")
plt.show()
````

**Exerc√≠cio 2**  
Crie um gr√°fico de pizza com a propor√ß√£o de vendas por `produto`.

````python
grupo.plot(kind = "pie", title = "Propor√ß√£o de vendas por produto", autopct = "%.2f%%")
plt.ylabel("")
plt.tight_layout()
plt.show()
````

**Exerc√≠cio 3**  
Crie um gr√°fico de linha com a soma de `quantidade` por m√™s.

````python
df["data_venda"] = pd.to_datetime(df["data_venda"], errors = "coerce")
df["mes"] = df["data_venda"].dt.to_period("M")

serie = df.groupby("mes")["valor_total"].sum()
serie.plot(kind = "line", title = "Vendas por m√™s", marker = "o")
plt.grid(True)
plt.xlabel("M√™s")
plt.ylabel("Quantidade")
````

**Exerc√≠cio 4**  
Crie um gr√°fico de dispers√£o entre `preco_unitario` e `valor_total`.

````python
df.plot(kind = "scatter", x = "preco_unitario", y = "valor_total", title = "Dispers√£o entre pre√ßo unit√°rio e valor total")
plt.xlabel("Pre√ßo Unit√°rio")
plt.ylabel("Valor Total")
plt.grid(True)
plt.show()
````

````python
# 1.Crie um gr√°fico de barras com o total de valor_total por produto.
grupo = df.groupby("produto")["valor_total"].sum()
grupo.plot(kind = "bar", title = "Total de valor_total por produto", color = "black")
plt.xlabel("Produto")
plt.ylabel("Valor Total")
plt.grid(True)
plt.show()
````

````python
#2.Crie um gr√°fico de linha mostrando a m√©dia de preco_unitario ao longo dos meses.
df["data_venda"] = pd.to_datetime(df["data_venda"], errors = "coerce")
df["mes"] = df["data_venda"].dt.to_period("M")
grupo = df.groupby("mes")["preco_unitario"].mean()
grupo.plot(kind = "line", title = "M√©dia de pre√ßo unit√°rio ao longo dos meses", marker = "o")
plt.grid(True)
plt.xlabel("M√™s")
plt.ylabel("Pre√ßo Unit√°rio")
plt.show()
````

````python
#3.Crie um gr√°fico de dispers√£o entre quantidade e preco_unitario.
df.plot(kind = "scatter", x = "quantidade", y = "preco_unitario", title = "Dispers√£o entre quantidade e pre√ßo unit√°rio")
plt.xlabel("Quantidade")
plt.ylabel("Pre√ßo Unit√°rio")
plt.grid(True)
plt.show()
````

````python
#4.Crie um gr√°fico de pizza com a participa√ß√£o de cada produto no total de valor_total.
grupo = df.groupby("produto")["valor_total"]
grupo.sum().plot(kind = "pie", title = "Participa√ß√£o de cada produto no total de valor total", autopct = "%.2f%%")
plt.ylabel("")
plt.tight_layout()
plt.show()
````
