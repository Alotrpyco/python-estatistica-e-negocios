## üî¢ Parte 1 ‚Äì Estat√≠sticas descritivas

````python
import pandas as pd

df = pd.read_csv("vendas_ausente.csv", sep=";", decimal=",")

df.head(30)

#df.describe()
````

````python
# coluna valor total

# c√°lculo da m√©dia (.mean())
print(f"M√©dia: {df['valor_total'].mean():.2f}")

# c√°lculo da mediana (.median())
print("Mediana:", df["valor_total"].median())

print("Mediana:", df[["valor_total", "quantidade"]].median())
````

````python
print("Desvio padr√£o:", df["valor_total"].std())
print("M√°ximo:", df["valor_total"].max())
print("M√≠nimo:", df["valor_total"].min())

print("Desvio padr√£o:", df[["valor_total", "quantidade"]].std())
````

````python
# estatisticas descritivas

df.describe()
````

````python
#quantis .quantile(numero que varia de 0 a 1)

print("Primeiro quartil do custo total {:.2f}".format(df['custo_total'].quantile(0.25)) )
````

## üîó Parte 2 ‚Äì Agrupando dados com `groupby()`

````python
#vendas m√©dias por produto
agrupado = df.groupby("produto")["valor_total"].mean()
print(agrupado)
````

````python
#vendas totais por produto
agrupadototais = df.groupby("produto")["valor_total"].sum()
print(agrupadototais)
````

````python
#totais por produto (para todas as colunas)
agrupadototais = df.groupby("produto").sum()
print(agrupadototais)
````

````python
contagem = df.groupby("produto").count() # numero de ocorr√™ncias n√£o vazias para cada coluna
print(contagem)

print(contagem[["quantidade"]])
````

## üìä Parte 3 ‚Äì Contagem de valores com `value_counts()`

````python
df["produto"].value_counts() #conta a frequencia de observa√ß√µes na coluna produto
````

## üßπ Parte 4 ‚Äì Lidando com valores ausentes

````python
valores = df["valor_total"].fillna(0)
print("M√©dia com ausentes tratados:", valores.mean())
print("M√©dia sem ausentes:", df["valor_total"].mean())
````
