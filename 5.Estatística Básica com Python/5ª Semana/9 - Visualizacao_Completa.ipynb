## ðŸ§° ImportaÃ§Ã£o de Bibliotecas

````python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('default')
````

## ðŸ“Š Parte 1 â€“ GrÃ¡fico de Barras (Bar Plot)

````python
df = pd.read_csv("vendas_ausente.csv", sep=";", decimal=",")

grupo = df.groupby("produto")["valor_total"].mean().sort_values()

#print(grupo)

grupo.plot(kind="bar", title="MÃ©dia de Receita por Produto", color="yellow")
plt.ylabel("Valor mÃ©dio")
plt.xlabel("Produto")
plt.tight_layout()
plt.show()
````

## ðŸ¥§ Parte 2 â€“ GrÃ¡fico de Pizza (Pie Plot)

````python
grupo.plot(kind="pie", autopct="%1.1f%%", title="ParticipaÃ§Ã£o no valor das Vendas")
plt.ylabel("")
plt.tight_layout()
plt.show()
````

````python
df["produto"].value_counts().plot(kind="pie", autopct="%1.1f%%", title="Percentual de ocorrÃªncias")
plt.ylabel("")
plt.tight_layout()
plt.show()
````

## ðŸ“ˆ Parte 3 â€“ GrÃ¡fico de Linha (Line Plot) para SÃ©ries Temporais

````python
df["data_venda"] = pd.to_datetime(df["data_venda"], errors='coerce')
df["mes"] = df["data_venda"].dt.to_period("M")

serie = df.groupby("mes")["valor_total"].sum()
serie.plot(kind="line", marker="o", title="Receita Total por MÃªs")
plt.ylabel("Receita")
plt.xlabel("MÃªs")
plt.grid(True)
plt.tight_layout()
plt.show()
````

## ðŸ”µ Parte 4 â€“ GrÃ¡fico de DispersÃ£o (Scatter Plot)

````python
df.plot(kind="scatter", x="quantidade", y="valor_total", title="RelaÃ§Ã£o entre Quantidade e Receita")
plt.grid(True)
plt.tight_layout()
plt.show()
````
