## 📊 Parte 5 – Exemplo prático: analisando vendas

```python
dados = pd.read_csv("vendas.csv", sep=";")
dados["lucro"] = dados["valor_total"] - dados["custo_total"]
print("Lucro total:", dados["lucro"].sum())
dados[dados["lucro"] > 500]
```
